```python
# Parity dataset

from circuits import parity

import random
random_number = random.randint(0, 1000)

input_seq_len=4
data  = parity(
    vocab_size=256,
    num_examples=10,
    input_seq_len=input_seq_len,
    seed=random_number,
)
x= data.inputs[0]
y= data.labels[0]

print(f"Lengths: {len(x)}, {len(y)}\n")
num_one = 0
for i, (_x, _y) in enumerate(zip(x, y)):
    if _x.item() == 1:
        num_one += 1
    print(_x.item(), _y.item())

print(num_one)
```
